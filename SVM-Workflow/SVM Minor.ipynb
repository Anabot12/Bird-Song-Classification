```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.svm import LinearSVC
from sklearn.multioutput import MultiOutputClassifier
from sklearn.multiclass import OneVsRestClassifier, OneVsOneClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, classification_report, r2_score, mean_squared_error, mean_absolute_error
```

```python
data_csv = pd.read_csv("E:\\Codes\\Python_codes\\data_sets\\minor audio csv\\combined_features.csv")
```

```python
data_csv
```

```python
feature_data = data_csv.iloc[:, 6:32]
feature_data
```

```python
label_data = data_csv.iloc[:, 1:3]
label_data
```

```python
feature_train, feature_test, label_train, label_test = train_test_split(feature_data, label_data, test_size=0.3, random_state=13)
```

```python
standard_scaler = StandardScaler()
feature_train = standard_scaler.fit_transform(feature_train)
feature_test = standard_scaler.transform(feature_test)
```

```python
support_vm = svm.SVC()
classifier = MultiOutputClassifier(support_vm, n_jobs=2)
# classifier = MultiOutputClassifier(n_jobs=2, estimator= LinearSVC())
# classifier = OneVsRestClassifier(estimator= LinearSVC)
# classifier = svm.SVC()
# classifier = OneVsOneClassifier(estimator= LinearSVC)
classifier.fit(feature_train, label_train)
```

```python
prediction_classifier = classifier.predict(feature_test)
print("Predicted output : \n", prediction_classifier)
```

```python
accuracy = np.mean(np.all(label_test == prediction_classifier, axis=1))
print("Accuracy of the model is : ",accuracy*100, "%")
```
